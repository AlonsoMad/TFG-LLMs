```python
import pandas as pd
import numpy as np
import os
path = '/export/usuarios_ml4ds/ammesa/Data/0_input_data'
fname = 'dataset_2025-05-17.parquet.gzip'
filepath = os.path.join(path, fname)
```

```python
df = pd.read_parquet(filepath); df
```

```python
df_es=df[df['lang']=='es']
df_en=df[df['lang']=='en']
```

```python
df_es = df_es[~((df_es['id'] <= 17999) & (df_es['id'] >= 16000))]
```

```python
df_en = df_en[~((df_en['id'] >= 8000) & (df_en['id'] <= 9999))]
```

```python
df_es['equivalence'] = np.where(df_es['id'] >= 18000, 0, 1)
```

```python
df_en['equivalence'] = np.where(df_en['id'] >= 6000, 0, 1)
```

```python
final_df = pd.concat([df_en, df_es])

len(final_df[final_df['equivalence']==1])/len(final_df)
```
