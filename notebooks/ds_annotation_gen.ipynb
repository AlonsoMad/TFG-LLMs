```python
def compute_doc_id_shares(df1, df2, df3, label):
    s1 = set(df1['doc_id'])
    s2 = set(df2['doc_id'])
    s3 = set(df3['doc_id'])

    intersection = s1 & s2 & s3
    union = s1 | s2 | s3

    print(f"--- {label} ---")
    print(f"Total in df1: {len(s1)}")
    print(f"Total in df2: {len(s2)}")
    print(f"Total in df3: {len(s3)}")
    print(f"Intersection count: {len(intersection)}")
    print(f"Share in df1: {len(intersection) / len(s1):.2%}")
    print(f"Share in df2: {len(intersection) / len(s2):.2%}")
    print(f"Share in df3: {len(intersection) / len(s3):.2%}")
    print(f"Share of intersection over union: {len(intersection) / len(union):.2%}")
    print()

# Group 1
compute_doc_id_shares(trans_32_df, match_10_df, zs_1_df, "trans_32 - match_10 - zs_1")

# Group 2
compute_doc_id_shares(trans_6_df, match_25_df, zs_7_df, "trans_6 - match_25 - zs_7")

# Group 3
compute_doc_id_shares(trans_31_df, match_48_df, zs_0_df, "trans_31 - match_48 - zs_0")
```
